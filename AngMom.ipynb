```python

fig, axs = plt.subplots(nrows= 1, ncols=4, figsize=(12, 4))

for ax, snap in zip(axs, [50, 75, 100, 119]):

    with h5.File(dir_name+files[0], "r") as hf:

        AngMom = hf[f"Snap_{snap:03d}/AngMom"][()]
        
        Mass_tot = np.array(hf[f"Snap_{snap:03d}/Mass_tot"]) * 1e10
    
        # AngMom =  AngMom [Mass_tot > 1e9]
        # Mass_tot =  Mass_tot [Mass_tot > 1e9]

        ax.hist2d(np.log10(Mass_tot), np.log10(AngMom), bins=100, density=1, cmap='Greys')

        # ax.set_xlim(8.9, 12)
        # ax.set_ylim(0,0.6)
        # ax.set_aspect("equal")
        ax.set_title(f"snap_{snap:03d}")
        ax.set_xlabel(f"log(Mass_tot)")



axs[0].set_ylabel("log(AngMom) stored")
        
fig.tight_layout(pad=0.1)    
fig.subplots_adjust(hspace=0.1, wspace=0.1)
```

```python

fig, axs = plt.subplots(nrows= 1, ncols=4, figsize=(12, 4))

for ax, snap in zip(axs, [50, 75, 100, 119]):
    
    with h5.File(dir_name+files[1], "r") as hf:
        
        Lx =  np.array(hf[f"Snap_{snap:03d}/Lx"])
        Ly =  np.array(hf[f"Snap_{snap:03d}/Ly"])
        Lz =  np.array(hf[f"Snap_{snap:03d}/Lz"])
    
        Mass_tot = np.array(hf["Snap_{:03d}/{}".format(snap, "Mass_tot")])
        AngMom = np.sqrt(Lx**2 + Ly**2 + Lz**2)/Mass_tot

        ax.hist2d(np.log10(Mass_tot*1e10), np.log10(AngMom), bins=100, density=1, cmap='Greys')

        # ax.set_xlim(8.9, 12)
        # ax.set_ylim(0,0.6)
        # ax.set_aspect("equal")
        ax.set_title(f"snap_{snap:03d}")
        ax.set_xlabel(f"log(Mass_tot)")



axs[0].set_ylabel("log(AngMom) computed")
        
fig.tight_layout(pad=0.1)    
fig.subplots_adjust(hspace=0.1, wspace=0.1)
```
